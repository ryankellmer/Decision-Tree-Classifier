```python
import pandas as pd #  Offers data manipulation, i will be using it for manipulating numerical tabels.
import matplotlib.pyplot as plt # Used for displaying data in figures and  graphs.
from sklearn.tree import DecisionTreeClassifier # tree structure that breaks down into smaller subsets resulting in a prediction. Traverse to find prediction
from sklearn.tree import plot_tree  #  adds onto matplotlib, allows for production of tree figures.
from sklearn.model_selection import train_test_split # Used to split data into our training and testing  data.
```

```python
# df is Data Frame, which will hold the tabel of data from our zoo.csv file.
df = pd.read_csv('zoo.csv')
df
```

```python
# We will show the first 5 rows of our dataframe.
df.head()
```

```python
# X will represent our Indepent variable or our attributes in our table. X will hold hair - catsize. Animal name is not a factor in the classification.
X = df.iloc[:, 1:17]
```

```python
# Show the first 5 instances of our new data set X. Notice class_type and animal name are no longer in the table.
X.head()
```

```python
# y will represent our dependent variable or our class type. this variable will only hold the class type which is numbered 1 - 7.
y = df.iloc[:, 17]
```

```python
# Show the first 5 instances of our y variable.
y.head()
```

```python
# 60% of our data will be used for testing our classifier and 40% to train our classifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.6)
```

```python
# our classifier dataframe will be a Decision Tree Classifier.
clf_df = DecisionTreeClassifier()
```

```python
# Were now training our Decision tree classifier with our training data so it can predict or evaluate the classifier.
clf_df = clf_df.fit(X_train, y_train)

# This compares our testing data to our trained classifier to see how accuratly it can predict the result.
print("Accuracy Score: ", clf_df.score(X_test, y_test))
```

```python
# plot figure size.
plt.figure(figsize=(15, 7.5))

# Print a tree graph where the classifier types are changed from numerical data to their equivalent catagorial names.
plot_tree(clf_df, filled= True, rounded= True, class_names= ["Mammal", "Bird", "Reptile", "Fish", "Amphibian", "Bug", "Invertebrate"], feature_names= X.columns );
```
